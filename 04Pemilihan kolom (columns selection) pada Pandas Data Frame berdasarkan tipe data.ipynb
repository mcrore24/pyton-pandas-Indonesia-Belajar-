```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#buatkolom 5
n_rows = 5

#buatbaris 2
n_cols = 2

cols = ['bil_pecahan', 'bil_bulat']


#membuat data frame dengan bilangan random integer dari angka 1 sampai 20 yang diisi didalam kolom rows dan cols
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                    columns=cols)

#mengubah tipe data awal integer ke tipe data float dengan fungsi astype
df['bil_pecahan']= df['bil_pecahan'].astype('float')

#membuat data index dengan tipe data date dan jumlah sesuai dengan variable n_rows
#dan freq berfungsi sebagai frekuensi dimana H berupa jam
df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')

#mereset dan membuat index baru berupa tipe int dari angka 0
df = df.reset_index()

#membuat kolom dengan set tipe data string ,value sesuai dengan yang ada pada fungsi list
df['teks'] = list('ABCDE')
df
```

```python
#memilih kolom dengan tipe data numerik (number)
df.select_dtypes(include='number')
```

```python
#memilih kolom dengan tipe data numerik (int)
df.select_dtypes(include='int')
```

```python
#memilih kolom dengan tipe data string atau object
df.select_dtypes(include='object')
```

```python
#memilih kolom dengan tipe data datetime
df.select_dtypes(include='datetime')
```

```python
#memilih kolom dengan kombinasi tipe data  string atau object dan datetime
#jika ingin menambah silahkan menambahkan sesuai tipe data yang diiginkan
df.select_dtypes(include=['datetime','object'])
```

```python
#mengurutkan bilangan pecahan dari kecil ke besar
df.sort_values(by=['bil_pecahan'], inplace=True)
df
```

```python
#mengurutkan bilangan pecahan sesuai dengan nilainya false atau true
df.sort_values(by=['bil_pecahan', 'bil_bulat'], ascending=[True, True])
df
```

```python
#mengurutkan bilangan pecahan dari kecil ke besar
df.sort_values(by=['bil_bulat'], ascending=False)
df
```
