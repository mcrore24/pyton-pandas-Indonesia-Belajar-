```python
import pandas as pd


print(pd.__version__)
```

```python
#dalam data frame sengaja di adakan ketidak konsistenan pada penamaan jenis kelamin dan juga kombinasi huruf besar dan kecil
df = pd.DataFrame({'jenis_kelamin': ['Pria', 'Wanita', 'lelaki', 'Lelaki', 'perempuan' ],
                   'usia': [23, 21, 24, 22, 21],
                   'shift': ['pagi', 'siang', 'Malam', 'Siang', 'pagi']})
df
```

```python
#MAPPING DATA FRAME DENGAN APPLYMAP ()
#apabila nilai dari varible x adalah string maka akan setiap data string akan dijadikan huruf kecil dengan fungsi lower() jika tidak maka akan dikembalikan seperti nilai awal
df = df.applymap(lambda x: x.lower() if type(x) == str else x)
df
```

```python
#APPLYMAP() dengan dicrionary
#malakukan standarisasi pada variable mapping
#applymap() berfungsi untuk mngubah nilai value ke baris pada seluruh kolon oleh karena itu nilai pada kolom usia menjadi non karena belum dilakukan mapping
mapping= {'pria':'L',
          'lelaki': 'L',
          'wanita': 'P',
          'perempuan': 'P',
          'pagi' :1,
          'siang':2,
          'malam':3}

df.applymap(mapping.get)
```

```python
#menampilkan data frame dengan value pada kolom jenis_kelamin dan shift yang dikenakan fungsung applymap pada variable value agar kolom usia tidak menjadi nilai none
df[['jenis_kelamin', 'shift']] = df[['jenis_kelamin', 'shift']].applymap(mapping.get)
df
```
