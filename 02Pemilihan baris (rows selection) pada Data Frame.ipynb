```python
import pandas as pd
import numpy as np
print(pd.__version__)
print(np.__version__)
```

```python
#buatkolom 10
n_rows = 10

#buatbaris 5
n_cols = 5

#menentukan judul colom
#tuple() mengkonversi data menjadi string / angka
cols = tuple('ABCDE')

#membuat data frame dengan bilangan random integer dari angka 1 sampai 5 yang diisi didalam kolom rows dan cols
df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)),
                    columns=cols)
df
```

```python
#operator logika | (OR / atau) berfungsi untuk menampilkan kolom A bernilai 1 atau 3 
df[(df[ 'A'] == 1) | (df['A'] == 3)]
```

```python
#menampilkan dan menyederhanakan operator | (Or / atau) dengan fungsi isin() pada kolom A dengan nilai 1 atau 3
df[df[ 'A'].isin([1,3])]
```

```python
#operator negasi (~) berfungsi menampilkan data yang tidak sama dengan nilai yang diset
df[~df[ 'A'].isin([1,3])]
```
