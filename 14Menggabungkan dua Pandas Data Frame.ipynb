```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#buatkolom 5
n_rows = 5

#buatbaris 5
n_cols = 5

#menentukan judul colom
cols = tuple('ABCDE')

#membuat data frame dengan bilangan random integer dari angka 1 sampai 20 yang diisi didalam kolom rows dan cols
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                    columns=cols)
df.head()
```

```python
#buat dataframe1 di dalam variable df1
df1 = df.copy(deep = True)
#df drop berfungsi untuk membuat baris yang ditentukan
df1 = df1.drop([1, 4])
df1
```

```python
#buat dataframe2 di dalam variable df2
df2 = df.copy(deep = True)
#df drop berfungsi untuk membuat baris yang ditentukan
df2 = df2.drop([0, 3])
df2
```

```python
#mmenggabungkan 2 data frame dengan fungsi merge
#inner berfungsi menggabungkan 2 data frame yang baris pada dua dataframenya sama
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

```python
#outer berfungsi megadopsi hanya 1 nilai pada baris dari 2 data frame yang sama
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```
