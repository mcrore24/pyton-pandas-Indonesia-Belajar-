```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#buatkolom 10
n_rows = 10

#buatbaris 1
n_cols = 1

#menentukan judul colom
cols = ('usia',)

#membuat data frame dengan bilangan random integer dari angka 1 sampai 99 yang diisi didalam kolom rows dan cols
df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)),
                    columns=cols)
df
```

```python
#pengelompokan nilai numerik ke dalam beberapa kategori menggunakan fungsi cut()
#bins befungsi untuk mengelompokkan data int dari dataframe sebelumnya
#labels memberikan nama kirteria sesuia usia yang sudah di set
df[ 'kelompok_usia'] = pd.cut(df['usia'],
                            bins = [0, 18, 65, 99],
                            labels = ['anak', 'dewasa', 'manula'])
df
```
