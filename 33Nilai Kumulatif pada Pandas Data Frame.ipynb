```python
import pandas as pd


print(pd.__version__)
```

```python

d =  {'pemain': ['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Aspe', 'Joni'],
      'goal': [2, 1 ,3 ,1 ,1 ,2 ,2 ,3]}
df = pd.DataFrame(d)
df
```

```python
#Menampilklan nilai kumulatif (cummulative)
#cumsun() berfungsi menjumlahkan nilai setiap baris dari index 0 sampi akhir
df['goal'].cumsum().to_frame()
```

```python
#sebagai contoh fungsi cumsum() pada index 2 jumlah_goal_kumulatif = 6, didapat dari nilai index 0 sampai ke 2 (2+1+3 = 6)
df['jumlah_goal_kumulatif'] = df['goal'].cumsum( )
df
```

```python
#fungsi group by adalah mengelompkan data pemain sebagai contoh fungsi penggabungan fungsi
#cumsum() dan grup by pada index 4 jumlah_goal_kumulatif_tiap_pemain (Budi) = 4, didapat dari nilai index 0,3,4 (2+1+1 = 4)
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
```

```python
# fungsi cummax() untuk meampilkan nilai tertinggi dari setiap gol dari index 0 - 7
df['cummax'] = df['goal'].cummax()
df
```

```python
# fungsi cummin() untuk meampilkan nilai terendah dari setiap gol dari index 0 - 7
df['cummin'] = df['goal'].cummin( )
df
```

```python
#cumprod berfungsi menampilkan hasil total perkalian 
#sebagai contoh pada index ke 3 dari perkalian index besar ke kecil maka nilai cumpround = 6 dimana 1*3 = 3, 3*1 = 3, 3 * 2 = 6
df['cumprod'] = df['goal'].cumprod()
df
```
