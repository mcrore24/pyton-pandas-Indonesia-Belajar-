```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#buatkolom 10
n_rows = 10

#buatbaris 5
n_cols = 5

#menentukan judul colom
#tuple() mengkonversi data menjadi string / angka
cols = tuple('ABCDE')

#membuat data frame dengan bilangan random integer dari angka 1 sampai 20 yang diisi didalam kolom rows dan cols
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                    columns=cols)
df
```

```python
df.shape 
```

```python
#membagi poposi dengan nilai 70 persen untuk data pertama dan data kedua di sisanya atau pada df 2
proporsi = 0.7
df_1 = df.sample(frac=proporsi)
#df2 = df.drop berfungsi untuk menghapus datafram berdasrkan df1 dan memasukan nilai index sisa
df_2 = df.drop(df_1. index)
print(f'df_1 shape: {df_1.shape}')
print( f'df_2 shape: {df_2.shape}')
```
