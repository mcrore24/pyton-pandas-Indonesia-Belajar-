```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#buatkolom 
n_rows = 40
#buatbaris
n_cols = 5
#menentukan judul colom
cols = tuple('ABCDE')

#membuat data frame dengan bilangan random integer dari angka 1 sampai 20 yang diisi didalam kolom rows dan cols
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                    columns=cols)
df.head()
```

```python
#line Plot
#panggil 5 data awal
df.head()
```

```python
#menampilkan grafik line dengan fungsi plot dengan data frame yang sudah dibuat
#tidak ada library matlotlib dari pandas yang di gunakan, hanya menggunakan fungsi plot bawaan dari pandas
df.plot(kind='line')
```

```python
#menampilkan grafik dengan fungsi plot dengan data kolom A dan B
df[['A', 'B']].plot(kind='line')
```

```python
#bari plot
df.head()
```

```python
#menampilkan grafik bar dengan fungsi plot dengan data frame yang sudah dibuat
df.plot(kind='bar')
```

```python
#menampilkan grafik dengan fungsi plot dengan data kolom A dan B
df[['A', 'B']].plot(kind='bar')
```

```python
#manampilkan grafik bar pada kolom a dan b dengan 5 data pertama
df[['A', 'B']].head().plot(kind='bar')
```

```python
#manampilkan grafik bar pada kolom a dan b dengan 5 data pertama secara horizontal
df[['A', 'B']].head().plot(kind='barh')
```
