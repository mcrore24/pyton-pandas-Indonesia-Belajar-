```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#buatkolom 10
n_rows = 10

#buatbaris 5
n_cols = 5

#menentukan judul colom
#tuple() mengkonversi data menjadi string / angka
cols = tuple('ABCDE')

#membuat data frame dengan bilangan random integer dari angka 1 sampai 20 yang diisi didalam kolom rows dan cols
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                    columns=cols)
df
```

```python
#seleksi kolomm dan baru menggunakan loc
#loc merupakan sebuah attribut yang dimiliki dalam data frame pandas
#bagian pertama untuk baris (index) dan bagian kedua untuk kolom
df.loc[[0,3,4], ['B','E']]
```

```python
#selesksi baris dengan kondisi oprator
#manampilkan baris yang kolom B lebih besar dari 10
df.loc[df['B'] > 10 , ['B','C','D']]
```

```python
#slicing data frame dengan loc
#loc 0 : 4 berungsi menampilkan baris dari index 0 ke index 4
#loc B D berfungsi menampilkan data kolom dari b - d
df.loc[0:4, 'B' : 'D']
```
