```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#buatkolom 
#24 merepresentasikan jumlah jam dalam sehari
#365 merepresntasikan jumlah hari dalm setahun
n_rows = 365 *24
#buatbaris 2
n_cols = 2
#menentukan judul colom
cols = ['col1', 'col2']

#membuat data frame dengan bilangan random integer dari angka 1 sampai 20 yang diisi didalam kolom rows dan cols
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                    columns=cols)

#mengisi index dengan data deret waktu
#freq='H' menunjukan interval yang ditampilkan perjam
df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df
```

```python
#resampling data debgan interval mounthly
#resample( ) berfungsi untuk mengolompokan ulang data sesuai dengan parmeter yang ditentukan dalam konteks ini buluan atau (M)

df.resample('M')['col1'].sum().to_frame()
```

```python
#resampling data debgan interval daily
#resample( ) berfungsi untuk mengolompokan ulang data sesuai dengan parmeter yang ditentukan dalam konteks ini hari atau (D)

df.resample('D')['col1'].sum().to_frame()
```
