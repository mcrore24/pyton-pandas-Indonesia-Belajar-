```python
import pandas as pd


print(pd.__version__)
```

```python
d ={'penjual' : ['bejo', 'tejo', 'wati', 'bejo', 'cecep', 'tejo', 'wati', 'bejo'],
    'barang' : ['monitor', 'monitor', 'keyboard', 'mouse', 'keyboard', 'monitor ', 'laptop', 'monitor']}
df = pd.DataFrame(d)
df
```

```python
#manampilkan data count tiap penjualan berdarkan data kolom penjualan
df['count_tiap_penjual'] = df.groupby('penjual').cumcount() + 1
df
```

```python
#manampilkan data count tiap penjualan berdarkan data kolom penjualan dan barang
df['count_pasangan_ kolom'] = df.groupby(['penjual', 'barang']).cumcount() + 1
df
```
