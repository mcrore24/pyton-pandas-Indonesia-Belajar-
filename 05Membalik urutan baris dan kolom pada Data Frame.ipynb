```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#buatkolom 5
n_rows = 5

#buatbaris 5
n_cols = 5

#menentukan judul colom
#tuple() mengkonversi data menjadi string / angka
cols = tuple('ABCDE')

#membuat data frame dengan bilangan random integer dari angka 1 sampai 5 yang diisi didalam kolom rows dan cols
df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)),
                    columns=cols)
df
```

```python
#membalikan urutan kolom dengan 
# tanda : untuk menagatur baris, tanda :: mengatur kolom dan nilai kolom
df.loc[:, ::-1]
```

```python
#membalikan urutan baris dengan 
df.loc[::-1]
```

```python
#mengatur ulang index dan membuat hanya nilai yang berubah tapi judul baris tidak
df.loc[::-1].reset_index(drop=True)
```
